```python
def Schaffer(nparray):
    result = 0.5 + ((np.sin(np.sqrt(np.sum(nparray**2))))**2 - 0.5) / ((1+0.001*np.sum(nparray**2))**2)
    return result
```

```python
pso0 = PSO_0(func=Schaffer, dim=10, pop=40, lb = [-100]*10, ub = [100]*10, max_iter=1000)
pso0.run()
pso1 = PSO_1(func=Schaffer, dim=10, pop=40, lb = [-100]*10, ub = [100]*10, max_iter=1000)
pso1.run()
pso2 = PSO_2(func=Schaffer, dim=10, pop=40, lb = [-100]*10, ub = [100]*10, max_iter=1000, mul=0.1)
pso2.run()
plt.plot(np.array(pso1.gbest_y_hist),'r--')
plt.plot(np.array(pso0.gbest_y_hist),'b--')
plt.plot(np.array(pso2.gbest_y_hist),'g--')
plt.show()
```

```python
import numpy as np
```

```python
def Sphere(nparray):
    result = 0
    for i in range(len(nparray)):
        result += nparray[i] ** 2
    return result
```

```python
def Rosenbrock(nparray):
    result = 0
    for i in range(len(nparray)-1):
        result += (100 * (nparray[i+1] - nparray[i]**2) + (nparray[i]-1)**2)
    return -result
```

```python
def Rastrigin(nparray):
    result = 0
    for i in range(len(nparray)):
        result += (nparray[i]**2 - 10*np.cos(2*np.pi*nparray[i]) + 10)
    return result
```
